### Playstation 2

```python
import re
pages = ['A%E2%80%93K','L%E2%80%93Z']
titles = []
developers = []
publishers = []
released = []

def ordertodate(string):
    return string[-2:]+'-'+string[4:6]+'-'+string[:4]
 
for page in pages:
    url = f'https://en.wikipedia.org/wiki/List_of_PlayStation_2_games_({page})'
    print(url)
    html = urlopen(url)
    soup = BeautifulSoup(html, 'html.parser')
    tables = soup.find_all('table')
    for table in tables:
        rows = table.find_all('tr')    
        for row in rows:
            cells = row.find_all('td')
            if len(cells) == 7:
                title = cells[0]
                titles.append(title.text.strip())

                developer = cells[1]
                developers.append(developer.text.strip())

                publisher = cells[2]
                publishers.append(publisher.text.strip())
                
                release = cells[3]
                released.append(ordertodate(re.sub('\D','',release.text.strip())))
```

```python
df1 = pd.DataFrame({'titles':titles,'released':released,'developers':developers,'publishers':publishers})
df1.to_csv('PS2.csv',index=False)
df1
```
