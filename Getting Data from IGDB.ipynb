```python
for title in titles:
    x = requests.post(gamesurl,headers = header,data = (f'search "{title}"; fields id,name,genres,storyline,summary;').encode('utf-8'))
    data = json.loads(x.text)
    for d in data:
        if 'name' not in d.keys():
            continue
        if d['name'] == title:
            title_info_dict[title] = d       
        
print('Done')
```

```python
def fetch_id(info):
    if 'id' not in info.keys():
        return ''
    return info['id']
def fetch_summary(info):
    if 'summary' not in info.keys():
        return ''
    return info['summary'].replace('\n','')
def fetch_storyline(info):
    if 'storyline' not in info.keys():
        return ''
    return info['storyline'].replace('\n','').replace('"','')
def fetch_genres(info):
    if 'genres' not in info.keys():
        return []
    return info['genres']
```

```python
def convert_to_tuple(genre):
    return genre.replace("[",'').replace("]",'').replace(" ",'')
```

```python
windows_games_df['genres1'] = windows_games_df['genres'].apply(convert_to_tuple)
```

```python
genres = windows_games_df['genres1'].value_counts().index
```

```python
genres_ids=[]

for genre in genres:
    x = genre.split(',')
    for g in x:
        if (g!= '') and (g not in genres_ids):
            genres_ids.append(g)
```

```python
len(genres_ids)
```

```python
genres_ids_names={}
for ids in genres_ids:
    x = requests.post(genresurl,headers = header,data = (f'fields *; where id = ({ids});').encode('utf-8'))
    data = json.loads(x.text)
    genre_data = data[0]
    if 'name' in genre_data.keys():
        genres_ids_names[ids] = genre_data['name']
```

```python
genres_ids_names
```

```python
windows_games_df['genres1'][0].split(',')
```

```python
def get_genres_name(ids):
    if ids=='':
        return ''
    genres = ids.split(',')
    g = ''
    for genre in genres:
        g+=genres_ids_names[genre]+','
    return g[:-1]
```

```python
get_genres_name(windows_games_df['genres1'][0])
```

```python
windows_games_df['genres']=windows_games_df['genres1'].apply(get_genres_name)
```
