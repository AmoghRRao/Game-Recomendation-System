```python
import pandas as pd
```

```python
df = pd.read_csv('Windows_info_complete.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.fillna("",inplace =True)
```

```python
df['summary'] = df['summary'].apply(lambda x: x.split())
df['storyline'] = df['storyline'].apply(lambda x: x.split())
df['genres'] = df['genres'].apply(lambda x: x.split(','))
```

```python
df['developers']=df['developers'].apply(lambda x: [x])
df['publishers']=df['publishers'].apply(lambda x: [x])
```

```python
df['genres']=df['genres'].apply(lambda x: [i.replace(' ','') for i in x])
df['developers']=df['developers'].apply(lambda x: [i.replace(' ','') for i in x])
df['publishers']=df['publishers'].apply(lambda x: [i.replace(' ','') for i in x])
```

```python
df.head()
```

```python
df['tags'] = df['summary']+df['storyline']+df['genres']+df['developers']+df['publishers']
```

```python
df.head()
```

```python
new_df = df[['id','titles','tags']]
new_df.head()
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:' '.join(x))
```

```python
new_df['tags']=new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df
```

```python
import nltk
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y= []
    for i in text.split():
        y.append(ps.stem(i))
        #print(y)
    return" ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features =5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
cv.get_feature_names()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
def get_recommendation(game_name):
    game_index = new_df[new_df['titles']==game_name].index[0]
    distance = similarity[game_index]
    game_list = sorted(list(enumerate(distance)),reverse = True,key = lambda x :x[1])[1:6]
    for i in game_list:
        print(new_df.iloc[i[0]].titles)
```

```python
get_recommendation('Grand Theft Auto III')
```
